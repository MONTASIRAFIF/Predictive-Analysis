```python
import matplotlib.pyplot as plt
import numpy as np

# Data
months = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]
speed_min = [3.5, 3.8, 4.0, 4.2, 4.0, 4.5, 5.0, 4.8, 4.5, 4.0, 3.8, 3.5]
speed_max = [4.0, 4.3, 4.5, 4.7, 4.5, 5.0, 5.5, 5.3, 5.0, 4.5, 4.3, 4.0]
power_min = [98, 133, 178, 207, 178, 250, 408, 358, 250, 178, 133, 98]
power_max = [178, 218, 273, 323, 273, 408, 573, 493, 408, 273, 218, 178]

# Bar chart for speed range
x = np.arange(len(months))
width = 0.4

fig, ax1 = plt.subplots(figsize=(12, 6))
bars_min = ax1.bar(x - width / 2, speed_min, width, label='Min Speed (m/s)', color='#DCDCDC')
bars_max = ax1.bar(x + width / 2, speed_max, width, label='Max Speed (m/s)', color='#A9A9A9')

# Add text labels on top of each bar for min and max speed
for i in range(len(months)):
    ax1.text(x[i] - width / 2, speed_min[i] + 0.05, f'{speed_min[i]:.1f}', ha='center', va='bottom', fontsize=10)
    ax1.text(x[i] + width / 2, speed_max[i] + 0.05, f'{speed_max[i]:.1f}', ha='center', va='bottom', fontsize=10)

# Labels and title for speed
ax1.set_ylabel('Wind Speed (m/s)', fontsize=12)
ax1.set_xlabel('Month', fontsize=12)
ax1.set_title('Rayleigh Forecasts for Chittagong (40m)', fontsize=14)
ax1.set_xticks(x)
ax1.set_xticklabels(months, rotation=45)
ax1.legend(loc='upper left')

# Line chart for power range
ax2 = ax1.twinx()
ax2.plot(x, power_min, label='Min Power (W/m²)', color='orange', marker='o', linestyle='-')
ax2.plot(x, power_max, label='Max Power (W/m²)', color='red', marker='o', linestyle='-')
ax2.set_ylabel('Power (W/m²)', fontsize=12)
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
