```python
import matplotlib.pyplot as plt
import numpy as np

# Monthly data for Rangpur district
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
sunshine_hours = [8.55, 9.41, 11.54, 12.39, 12.43, 12.15, 11.69, 11.81, 11.22, 9.19, 8.62, 8.54]
solar_potential = [5.5 * 1.7 * 0.18 * hours for hours in sunshine_hours]  # kWh for each month

# Bar width and position adjustment
bar_width = 0.3
index = np.arange(len(months))

# Create figure and axis objects
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot the sunshine hours as bars
bars = ax1.bar(index, sunshine_hours, bar_width, label='Sunshine Hours (hrs)', color='orange')

# Add values above the bars
for bar in bars:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width() / 2, yval + 0.1, f'{yval:.2f}', ha='center', color='black', fontsize=10)

# Create a second y-axis for solar potential
ax2 = ax1.twinx()

# Plot the solar potential as a line
ax2.plot(index, solar_potential, label='Solar Energy Potential (kWh)', color='green', marker='o', linestyle='-', linewidth=2)

# Formatting the graph
ax1.set_xlabel('Month')
ax1.set_ylabel('Sunshine Hours (hrs)', color='orange')
ax2.set_ylabel('Solar Energy Potential (kWh)', color='green')
ax1.set_title('Monthly Sunshine Hours and Solar Energy Potential in Rangpur District')
ax1.set_xticks(index)
ax1.set_xticklabels(months)
ax1.tick_params(axis='y', labelcolor='orange')
ax2.tick_params(axis='y', labelcolor='green')

# Add legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the graph
plt.tight_layout()
plt.show()
```
